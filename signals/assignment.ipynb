```python
# %pip install matplotlib scipy
import signal
from matplotlib import pyplot as plt
import numpy as np
from scipy import signal
```

```python



x = np.arange(0,10,0.01)
x2 = np.linspace(0,1,1000,endpoint = True)
s=  np.sin(x)
c = np.cos(x)
saw = signal.sawtooth(10*x2*np.pi)
square = signal.square(10*x2*np.pi)
trig = signal.sawtooth(10*x2*np.pi,0.5)


figure, axis = plt.subplots(2,3,figsize = (20,10))

axis[0,0].plot(x,s)
axis[0,0].set_title("sin(t)",fontstyle = 'italic')
axis[0,0].axhline(y=0,color='k')

axis[0,1].plot(x,c)
axis[0,1].set_title("cos(x)",fontstyle = 'italic')
axis[0,2].plot(x2,saw)
axis[0,2].set_title("sawtooth",fontstyle = 'italic')
axis[1,0].plot(x2,square)
axis[1,0].set_title("square",fontstyle = 'italic')
axis[1,1].plot(x2,trig)
axis[1,1].set_title("triangle",fontstyle = 'italic')
plt.delaxes(axis[1,2])
for i in axis:
    for j in i:
        j.set_xlabel('x')

plt.show()
```
