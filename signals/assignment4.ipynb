```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import scipy.fftpack as fft
```

```python
search_pattern = [6,1,1,2]
input_signal = [6,8,6,6,8,10,8,6,1,1,2,4,4,4,4,6,5,7,8,5,2,3,5,6,7,8,8,6,5,5,5]
# #reverse the input signal
# input_signal = input_signal[::-1]
# #make the signal negative
# input_signal = -np.array(input_signal)



#convolve the input signal and search pattern
output_signal = sig.correlate(input_signal,search_pattern)
#make the output signal size equal to input signal size
output_signal = output_signal[:len(input_signal)]

#normalise output signal


#find the peak in the output signal and store the index
peak  = output_signal.argmax()
print(peak)



#plot the input search and output signal
plt.figure()
#increase the size of the figure
plt.rcParams["figure.figsize"] = (20,20)
plt.subplot(3,1,1)
plt.plot(input_signal)
plt.title('Input Signal')
plt.subplot(3,1,2)
plt.plot(search_pattern)
plt.title('Search Pattern')
plt.subplot(3,1,3)
plt.plot(output_signal)
plt.title('Output Signal')
plt.show()
```
