```python
# Ex. 2) a) Impulse

def impulse(a,y):
    delta =[]
    for sample in y:
        if sample == a:
            delta.append(1)
        else:
            delta.append(0)
    
    return delta

time = np.arange(-10,10,1);
amp = impulse(0,time)
plt.stem(time,amp)
plt.title('Impulse')
plt.xlabel('Time')
plt.ylabel('Amplitude = impulse(time)')
plt.axhline(y=0,color='r')
plt.show()
```

```python
# Ex. 2) b) Unit Step

def unitstep(a,y):
    delta =[]
    for sample in y:
        if sample >= a:
            delta.append(1)
        else:
            delta.append(0)
    
    return delta

time = np.arange(-10,10,1);
amp = unitstep(0,time)
plt.stem(time,amp)
plt.title('Unit Step')
plt.xlabel('Time')
plt.ylabel('Amplitude = unitstep(time)')
plt.axhline(y=0,color='r')
plt.show()
```

```python
# Ex. 2) c) Ramp

def ramp(a,y):
    delta =[]
    for sample in y:
        if sample >= a:
            delta.append(sample)
        else:
            delta.append(0)
    
    return delta

time = np.arange(-10,10,1);
amp = ramp(0,time)
plt.stem(time,amp)
plt.title('Ramp')
plt.xlabel('Time')
plt.ylabel('Amplitude = ramp(time)')
plt.axhline(y=0,color='r')
plt.show()
```

```python
# Ex. 2) d) Exponential Growth



time = np.arange(0,2,0.1);
amp = np.exp(time)
plt.stem(time,amp)
plt.title('Exponential Growth')
plt.xlabel('Time')
plt.ylabel('Amplitude = expgrowth(time)')
plt.axhline(y=0,color='r')
plt.show()
```

```python
# Ex. 2) e) Exponential Decay


time = np.arange(0,2,0.1);
amp = np.exp(-time)
plt.stem(time,amp)
plt.title('Exponential Decay')
plt.xlabel('Time')
plt.ylabel('Amplitude = expdecay(time)')
plt.axhline(y=0,color='r')
plt.show()

```

```python
# Ex. 4) a) u[n] = x[n] + y[n]

area = np.arange(-3,4,1);
x = np.array([3,-2,0,1,4,5,2])
y = np.array([0,7,1,-3,4,9,-2])
u = np.add(x,y)
plt.stem(area,u)
plt.title('u[n] = x[n] + y[n]')
plt.xlabel('n')
plt.show()
```

```python

# Ex. 4) b) v[n] = x[n].w[n]

area = np.arange(-3,4,1);
x = np.array([3,-2,0,1,4,5,2])
w = np.array([-5,4,3,6,-5,0,1])
v = np.multiply(x,w)
plt.stem(area,v)
plt.title('v[n] = x[n].w[n]')
plt.xlabel('n')
plt.show()

```

```python
# Ex. 4) c) s[n] = y[n] – w[n]

area = np.arange(-3,4,1);
y = np.array([0,7,1,-3,4,9,-2])
w = np.array([-5,4,3,6,-5,0,1])
s = np.subtract(y,w)
plt.stem(area,s)
plt.title('s[n] = y[n] – w[n]')
plt.xlabel('n')
plt.show()
```
