```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# search pattern = 6,8,10,8,6
# input= 6,8,6,6,8,10,8,6,4,3,2,4,4,4,4,4,4,4,1,1,
#  1,3,1,2,4,4,4,4,4,4,4,4,4,6,8,10,8,6,4,4

search_pattern = [6,8,10,8,6]
input = [6,8,6,6,8,10,8,6,4,3,2,4,4,4,4,4,4,4,1,1,1,3,1,2,4,4,4,4,4,4,4,4,4,6,8,10,8,6,4,4]


#finding the pattern first occurence using correlation

output = np.correlate(input, search_pattern, mode='full')

print(output)

#plot the input and output abd search pattern separately subplots

plt.figure(figsize=(5,5))
plt.subplot(3,1,1)
#increasing the size of the subplot to 10x10
plt.plot(input)
plt.title('Input')
#spacing between the subplots
plt.subplots_adjust(hspace=0.5) 
plt.subplot(3,1,2)
plt.plot(search_pattern)
plt.title('Search Pattern')
plt.subplots_adjust(hspace=0.5)
plt.subplot(3,1,3)
plt.plot(output)
plt.title('Output')
plt.show()

#find the peak value in the output
peak = np.argmax(output)

# get the starting index of the pattern
start = peak - len(search_pattern) + 2
print('First occurence of the pattern is at index: ', start)

#get all the peaks in the output
peaks = np.where(output == np.amax(output))[0]
# print(peaks)

#find the starting index of the pattern for all the peaks
start = peaks - len(search_pattern) + 2
print('Starting index of the pattern for all the peaks: ', start)




```

```python
import scipy.signal as sig


noisy_input = input + np.random.normal(0, 0.1, len(input))

filtered_input = noisy_input




#finding the pattern first occurence using correlation
output = np.correlate(filtered_input, search_pattern, mode='full')

#plot the input and output abd search pattern separately subplots

plt.figure(figsize=(5,5))
plt.subplot(3,1,1)
#increasing the size of the subplot to 10x10
plt.plot(noisy_input)
plt.title('Noisy Input')
#spacing between the subplots
plt.subplots_adjust(hspace=0.5)
plt.subplot(3,1,2)
plt.plot(search_pattern)
plt.title('Search Pattern')
plt.subplots_adjust(hspace=0.5)
plt.subplot(3,1,3)
plt.plot(output)
plt.title('Output')
plt.show()

#find the peak value in the output
peak = np.argmax(output)

# get the starting index of the pattern
start = peak - len(search_pattern) + 2
print('First occurence of the pattern is at index: ', start)


#get all the peaks in the output which are close to the peak value by 1
peaks = np.where(output >= np.amax(output)-5)[0]

# print(peaks)

#find the starting index of the pattern for all the peaks
start = peaks - len(search_pattern) + 2

print('Starting index of the pattern for all the peaks: ', start)

```
